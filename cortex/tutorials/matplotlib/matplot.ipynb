```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Single plot graphs

### - create a dummy data for plot experimentation

```python
appl_price = [93.23, 84.23, 32.34, 76.34, 45.46]
ms_price = [39.23, 85.32, 45.24, 56.43, 87.34]
year = [2014, 2015, 2016, 2017, 2018]
```

```python
plt.plot(year, appl_price)
plt.xlabel('Year')
plt.ylabel('stock price')
plt.show()
```

### - Adding Multiple plots in a single graph

```python
plt.plot(year, appl_price)
plt.plot(year, ms_price)
plt.xlabel('Year')
plt.ylabel('stock price')
plt.show()
```

### - Adding scatter plots

```python
plt.scatter(year, appl_price)
plt.scatter(year, ms_price)
plt.xlabel('Year')
plt.ylabel('stock price')
plt.show()
```

### - Plotting in a single line with different colors and patterns
( https://matplotlib.org/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D )

```python
plt.plot(year, appl_price, ':k',
         year, ms_price, '--g')
plt.xlabel('Year')
plt.ylabel('stock price')
plt.show()
```

### - Setting the plot window range

```python
plt.plot(year, appl_price, ':k',
         year, ms_price, '--g')
plt.xlabel('Year')
plt.ylabel('stock price')

plt.axis([2013, 2019, 20, 100]) # [xmin, xmax, ymin, ymax]

plt.show()
```

## Multi plot graphs

```python
fig_1 = plt.figure(1, figsize=(6.4, 4.8)) # width, height

chart_1 = fig_1.add_subplot(121) # 1 2 --> 1 row 2 columns, 1 --> 1st chart position
chart_2 = fig_1.add_subplot(122) # 1 2 --> 1 row 2 columns, 2 --> 2nd position in the chart
```

### Add data to these empty charts

```python
fig_1 = plt.figure(1, figsize=(6.4, 4.8)) # width, height

chart_1 = fig_1.add_subplot(121)
chart_2 = fig_1.add_subplot(122)

chart_1.plot(year, appl_price)
chart_2.scatter(year, ms_price)

plt.show()
```

```python
fig_1 = plt.figure(1, figsize=(6.4, 4.8)) # width, height

chart_1 = fig_1.add_subplot(221) #  2 x 2 grid, 1st position
chart_2 = fig_1.add_subplot(224) #  2 x 2 grid, 4th position

chart_1.plot(year, appl_price)
chart_2.scatter(year, ms_price)

plt.show()
```

## Setting tick locations and axes formating

```python
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib.ticker import MaxNLocator 
```

```python
appl_price = [93.23, 84.23, 32.34, 76.34, 45.46]
ms_price = [39.23, 85.32, 45.24, 56.43, 87.34]
year = [2014, 2015, 2016, 2017, 2018]
```

```python
fig_1 = plt.figure(1, figsize=(6.4, 4.8)) # width, height

chart_1 = fig_1.add_subplot(221) #  2 x 2 grid, 1st position
chart_1.xaxis.set_major_locator(MaxNLocator(integer=True))

chart_2 = fig_1.add_subplot(224) #  2 x 2 grid, 4th position
chart_2.xaxis.set_major_locator(MaxNLocator(integer=True))

chart_1.plot(year, appl_price)
chart_2.scatter(year, ms_price)

plt.show()
```

## consolidating all with subplots
